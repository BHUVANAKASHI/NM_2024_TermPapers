```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reset_rate_rtp_pos_45,avg_time1=np.loadtxt("rtp_pos_45.dat",unpack=True)
reset_rate_rtp_pos_5,avg_time1=np.loadtxt("rtp_pos_5.dat",unpack=True)
reset_rate_rtp_pos_55,avg_time2=np.loadtxt("rtp_pos_55.dat",unpack=True)
#reset_rate_rtp_neg_6,avg_time1=np.loadtxt("rtp_neg_6.dat",unpack=True)
reset_rate_rtp_neg_55,avg_time3=np.loadtxt("rtp_neg_5.5.dat",unpack=True)
reset_rate_rtp_neg_5,avg_time4=np.loadtxt("rtp_neg_5.dat",unpack=True)
```

```python
#plt.plot(reset_rate_rtp_pos_45,avg_time1,label="initial position=4.5")
plt.plot(reset_rate_rtp_pos_5,avg_time1,label="$x_0$=5,$v_0$=1")
plt.plot(reset_rate_rtp_pos_55,avg_time2,label="$x_0$=5.5,$v_0$=1")
#plt.plot(reset_rate_rtp_neg_6,avg_time1,label="initial position=6")
plt.plot(reset_rate_rtp_neg_55,avg_time3,label="$x_0$=5.5,$v_0$=-1")
plt.plot(reset_rate_rtp_neg_5,avg_time4,label="$x_0$=5,$v_0$=-1")
plt.ylim(0,100)
#plt.xlim(0,2)
plt.xlabel("Reset rate")
plt.ylabel("Average time for search")
plt.title("Free RTP with target at x=6.2")
plt.legend()
```
